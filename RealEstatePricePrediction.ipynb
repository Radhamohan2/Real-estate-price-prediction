```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('Real estate.csv')
data
```

```python
X=data.iloc[:,1:7] # taking the params as X1 X2 X3 X4
Y=data.iloc[:,-1] # house price of unit area
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
X_train
```

```python
sns.heatmap(X_train.corr(), annot = True);
```

```python
X_train.describe()
```

```python
#droping the latitude and longitude due to less relation

X_train = X_train.drop(["X5 latitude","X6 longitude"], axis = 1) 
X_test = X_test.drop(["X5 latitude","X6 longitude"], axis = 1)
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python
Y_pred=reg.predict(X_test)
Y_pred
```

```python
acc = reg.score(X_test, Y_test)*100
print(f"The accurancy of the model is {round(acc, 2)}%")
```

```python
import sklearn.metrics as sm
print("R2 score =", round(sm.r2_score(Y_test, Y_pred), 1))
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, Y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, Y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test, Y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, Y_pred), 2))
```
